```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import folium
import streamlit as st
```

# Load Dataset

```python
df = pd.read_csv("DataSet/covid_19_clean_complete.csv")
print(df.head())
```

# Data Cleaning & Preprocessing

```python
# Convert date
df['Date'] = pd.to_datetime(df['Date'])

# Fill missing values
df.fillna(0, inplace=True)

# Create Active Cases
df['Active'] = df['Confirmed'] - df['Recovered'] - df['Deaths']
```

# Aggregate City-Level Data

```python
city_data = df.groupby('City').sum().reset_index()
print(city_data)
```

# Seaborn Visualization

Box Plot

```python
sns.boxplot(x='City', y='Confirmed', data=df)
plt.xticks(rotation=90)
plt.title("Confirmed Cases Distribution")
plt.show()
```

Violin Plot

```python
sns.violinplot(x='City', y='Active', data=df)
plt.xticks(rotation=90)
plt.title("Active Cases Density")
plt.show()
```

Heatmap

```python
corr = df[['Confirmed','Recovered','Deaths','Active']].corr()

sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

# Plotly Interactive Charts

Line Chart

```python
fig = px.line(df, x='Date', y='Confirmed', color='City',
              title="COVID Trend")
fig.show()
```

Pie Chart

```python
fig = px.pie(city_data, names='City', values='Confirmed',
             title="City-wise Cases")
fig.show()
```

# Folium Map Visualization

```python
m = folium.Map(location=[22.5, 78.9], zoom_start=5)

for i in range(len(city_data)):
    folium.CircleMarker(
        location=[city_data.iloc[i]['Latitude'], city_data.iloc[i]['Longitude']],
        radius=city_data.iloc[i]['Confirmed'] / 1000,
        popup=city_data.iloc[i]['City'],
        color='red',
        fill=True
    ).add_to(m)

m.save("covid_map.html")
```

# Create Dashboard using Streamlit

```python
df = pd.read_csv("covid_city_data.csv")

st.title("COVID-19 Dashboard")

fig = px.line(df, x='Date', y='Confirmed', color='City')
st.plotly_chart(fig)

fig2 = px.pie(df, names='City', values='Confirmed')
st.plotly_chart(fig2)
```

```python
city = st.selectbox("Select City", df['City'].unique())

filtered = df[df['City'] == city]

fig = px.line(filtered, x='Date', y='Confirmed')
st.plotly_chart(fig)
```
